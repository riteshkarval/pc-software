```python
import cv2
import json, os
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

## Function to load images and corresponding labels

```python
def read_data_from_dir(datadir, grayscale=True):
    imagedir = os.path.join(datadir, "IMAGES")
    image_files = list()
    label_files = list()
    for file_type in (".png", ".jpg"):
        image_files.extend(glob.glob(os.path.join(imagedir, "**/*" + file_type), recursive=True))
    if len(image_files) == 0:
        return None
    images = []
    for each_image_file in image_files:
        if grayscale:
            img = cv2.imread(each_image_file, cv2.IMREAD_GRAYSCALE)
        else:
            img = cv2.imread(each_image_file)
        if img is not None:
            images.append(img)
            filebasename, _ = os.path.splitext(each_image_file)
            label_file_name = filebasename.replace("IMAGES", "TEXT_LABELS") + ".gui"
            with open(label_file_name) as f:
                label_files.append(f.read())
    images = np.asarray(images)
    return images, label_files
```

## Loading dataset D1

```python
d1_images, d1_label_files = read_data_from_dir("../D1", grayscale=False)
```

```python
len(d1_images), len(d1_label_files)
```

## Visualizing image and corresponding label

```python
plt.figure(figsize = (9,6))
plt.axis('off')
plt.imshow(d1_images[0], interpolation='nearest', aspect='auto')
```
